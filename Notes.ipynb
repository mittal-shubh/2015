### Queries which can be performed on bytes or byte-like objects (like source1):
- Is the word 'Alice' mentioned on the beautiful soup homepage?
<**Sol:** `Encode the word into the format same as that of parsed html-page. And then using` **`in`** `in-built method of python.`>
- How often does the word 'Soup' occur on the site?
<**Hint:** use  `.count()`>
- At what index occurs the substring 'alien video games' ?
<**Hint:** use `.find()`>

```python
## is 'Alice' in source?
print(b'Alice' in source1)

## count occurences of 'Soup'
print(source1.count(b'Soup'))

## find index of 'alien video games'
position = source1.find(b'a lot of time')
print(position)

## quick test to see the substring in the source variable
## you can access strings like lists
print(source1[position:position + 17])

## or the tidier version:
print(source1[position:position + len(b'a lot of time')])
```

Beautiful Soup is a Python library designed for quick turnaround projects like screen-scraping.

Beautiful Soup parses anything you give it, and does the tree traversal stuff for you. You can tell it "Find all the links", or "Find all the links of class externalLink", or "Find all the links whose urls match "foo.com", or "Find the table heading that's got bold text, then give me that text."

Valuable data that was once locked up in poorly-designed websites is now within your reach. Projects that would have taken hours take only minutes with Beautiful Soup.

```python
article_list = []
for i in range(305914,305714,-1):
    url = "https://www.entrepreneur.com/article/%d" %i
    try:
        r = urllib.request.urlopen(url)
    except HTTPError:
        continue
    else:
        #print(i)
        data = r.read()
        soup = bs4.BeautifulSoup(data, "lxml")
        article = {}
        article["date"] = soup.find("meta", attrs={"name":"sailthru.date"})["content"].split("Z")[0].split("T")[0]
        article["time"] = soup.find("meta", attrs={"name":"sailthru.date"})["content"].split("Z")[0].split("T")[1]
        article['language'] = soup.find('html')['lang']
        article["edition"] = soup.find("meta", attrs={"name":"contentEdition"})['content']
        article["tags"] = soup.find("meta", attrs={"name":"sailthru.tags"})['content'].split(",")
        article["link"] = soup.find("meta", attrs={"name":"original-source"})['content']
        try: int(soup.find("meta", attrs={"name":"original-source"})['content'].split("/")[-1])
        except ValueError:
            article["id"] = int(soup.find(class_="qs shareme ga-click")["data-ga-label"].split("-")[1])
        else:
            article["id"] = int(soup.find("meta", attrs={"name":"original-source"})['content'].split("/")[-1])
        article["title"] = soup.find('title').get_text()
        article["author"] = soup.find("meta", attrs={"name":"sailthru.author"})['content']
        try: soup.select('i[class="icon-home"]')[0].find_parent("a")['href']
        except IndexError: 
            home_id = "NA"
        else: 
            home_id = soup.select('i[class="icon-home"]')[0].find_parent("a")['href']
        article["author ids"] = [home_id,
                                 soup.select('a[href^="https://www.linkedin.com/"]')[0]['href']]
        for tag in soup.select('a[href^="https://www.entrepreneur.com/article"]'):
            if tag.find_parent('p') is not None and (tag.find_parent('p').find("strong") is not None or 
                                                     tag.find_parent('p').find("b") is not None):
                tag1 = tag.find_parent("p")
                tag1.name = "adlink"
        soup_p = []
        for i in range(len(soup.find_all(class_="a-s pd articletext"))):
            soup_p.append(soup.find_all(class_ = "a-s pd articletext")[i].find_all(['p','h2']))
        text=[]
        for i in range(len(soup_p)):
            for p in soup_p[i]:
                text.append(p.get_text(strip=True))
        "".join(text)
        article["content"] = text
        article_list.append(article)
```
